```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import random

# Load dataset
df = pd.read_csv(r'C:\Users\venka\OneDrive\Desktop\Projects-20240722T093004Z-001\Projects\mobile_phone_pricing\Mobile Phone Pricing\dataset.csv')

# Split features and target
X = df.iloc[:, :-1]
y = df.iloc[:, -1]

# Divide each column into 4 parts: low, medium, high, very high
binned = pd.DataFrame()
for col in X.columns:
    try:
        binned[col] = pd.cut(X[col], bins=4, labels=['low', 'medium', 'high', 'very high'])
    except Exception:
        binned[col] = 'medium'

# Encode categorical bins numerically
X_encoded = binned.apply(lambda x: x.astype('category').cat.codes)

# Split dataset for training and testing
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

# Train Decision Tree model
model = DecisionTreeClassifier(criterion='entropy', max_depth=6, random_state=42)
model.fit(X_train, y_train)

# Model accuracy
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

# Predict for 20 random rows (new 20 on each run)
sample_indices = random.sample(range(len(df)), 20)
samples = X.loc[sample_indices]
sample_y_true = y.loc[sample_indices]

sample_binned = pd.DataFrame()
for col in samples.columns:
    sample_binned[col] = pd.cut(samples[col], bins=4, labels=['low', 'medium', 'high', 'very high'])

# Encode sample data
sample_encoded = sample_binned.apply(lambda x: x.astype('category').cat.codes)
sample_preds = model.predict(sample_encoded)

# Map numerical prices to readable labels
map_price = {0: 'low', 1: 'medium', 2: 'high', 3: 'very high'}
sample_actual_text = sample_y_true.map(map_price).values
sample_pred_text = [map_price[p] for p in sample_preds]

# Combine columns for output display
output = pd.concat([
    samples.iloc[:, :5].reset_index(drop=True),
    pd.Series(sample_actual_text, name='Actual Price'),
    pd.Series(sample_pred_text, name='Predicted Price')
], axis=1)

# Print accuracy and prediction results
print(f'Accuracy: {accuracy:.2f}\n')
print('Sample Predictions for 20 Random Rows (Actual vs Predicted):\n')
print(output.to_string(index=False))
```
